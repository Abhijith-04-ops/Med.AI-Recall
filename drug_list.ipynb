```python
import pandas as pd
import numpy as np
import csv
import re
```

```python
df = pd.read_csv('drug_name_list.csv')
```

```python
df.head()
```

```python
data = df['name'].dropna().str.lower().str.strip().tolist()
```

```python
drug_list=[]

for i in data:
    #cleaned_name = re.sub(r'\s*(tablet|syrup|inhaler|capsule|suspension|injection|spray|drops|cream|liquid)\b', '', i, flags=re.IGNORECASE)
    drug_list.append(i)
```

```python
alter_data = pd.read_csv('dataset01.csv')
alter_data = alter_data.drop(columns=['Unnamed: 0'])
alter_data.head()

# Filter rows where 'name' column contains any of the drug forms
#pattern = r'\b(tablet|syrup|inhaler|capsule|suspension|injection|spray|drops|cream|liquid)\b'
#alter_data['name'] = alter_data['name'].str.replace(pattern, '', regex=True)
#alter_data['name'] = alter_data['name'].str.strip().str.replace(r'\s+', ' ', regex=True)
# Save the filtered data to a new CSV
alter_data.to_csv('alternate_drugs.csv', index=False)
```

```python
alter_drug={}
with open('alternate_drugs.csv', mode='r',encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  
    for row in reader:
        key = row[0]  
        value = row[1]  
        alter_drug[key] = value
```

```python
dt = list(alter_drug.keys())
dt
```

```python
from rapidfuzz import process, fuzz

def correct_drug_name(user_query, drug_list, threshold=75):
    user_query = user_query.lower().strip()
    match, score, _ = process.extractOne(user_query, drug_list, scorer=fuzz.ratio)
    if score >= threshold:
        return match
    else:
        return None

# Example usage
   
user_input = input('Enter the drug name : ')
corrected = correct_drug_name(user_input, drug_list)
print(f'Did you mean {corrected} ?')
choice = input('[Y/N]')

if choice.lower() == 'y':
    user_drug = corrected
    if user_drug in list(alter_drug.keys()):
        new_drugs = alter_drug[user_drug]
        print(new_drugs)
    else:    
        print('No alternate drugs found for:', user_drug)
else:
    print('End')
```
